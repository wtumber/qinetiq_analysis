```python
BAE = yf.Ticker("BA.L")
BAE_hist = BAE.history(start="2019-01-01",interval="1d").reset_index()
BOE = yf.Ticker("BA")
BOE_hist = BOE.history(start="2019-01-01",interval="1d").reset_index()
RR = yf.Ticker("RR.L")
RR_hist = RR.history(start="2019-01-01",interval="1d").reset_index()
CHG = yf.Ticker("CHG.L")
CHG_hist = CHG.history(start="2019-01-01",interval="1d").reset_index()
LMT = yf.Ticker("LMT")
LMT_hist = LMT.history(start="2019-01-01",interval="1d").reset_index()
ULE = yf.Ticker("ULE.L")
ULE_hist = ULE.history(start="2019-01-01",interval="1d").reset_index()
QQ_2year = yf.Ticker("QQ.L").history(start="2019-01-01",interval="1d").reset_index()
QQ1 = yf.Ticker("QQ.L").history(start="2020-01-01",end="2021-01-01",interval="1d").reset_index()
FTSE = yf.Ticker("^FTMC").history(start="2020-01-01",end="2021-01-01",interval="1d").reset_index()
FTAS = yf.Ticker("^FTAS").history(start="2020-01-01",end="2021-01-01",interval="1d").reset_index()#.rolling(window=3).mean()
```

```python
tickers = [BAE_hist,BOE_hist,RR_hist,CHG_hist,LMT_hist,ULE_hist,QQ_2year,QQ1,FTAS,FTSE]

for ticker in tickers:
    ticker["Date"] = pd.to_datetime(ticker['Date'], infer_datetime_format=True)
    ticker["prct"] = ((ticker['Close']-ticker['Close'][0])/ticker['Close'][0])*100
```
